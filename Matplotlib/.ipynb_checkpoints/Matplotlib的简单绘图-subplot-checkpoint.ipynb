```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 5.0)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
len(x)
```

```python
plt.plot(x, y1, 'b--', label='sin(πx)')
plt.ylabel('y1 value')
plt.plot(x, y2, 'r--', label='sin(2πx)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y demox')
plt.legend()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(2, 1, 2)
plt.plot(x, y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x value')
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x value')
```

```python
plt.subplot(221)
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(222)
plt.plot(x, y2, 'r--')
plt.ylabel('y2')
plt.xlabel('x value')
plt.subplot(223)
plt.plot(x, y1, 'r*')
plt.subplot(224)
plt.plot(x, y1, 'b*')
```
