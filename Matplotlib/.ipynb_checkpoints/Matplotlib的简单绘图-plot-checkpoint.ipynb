```python
t = np.arange(0.0, 2.0, 0.1)
t.size
```

```python
s = np.sin(t*np.pi)
```

```python
plt.plot(t, s, 'r--', t*2, s, '--')
plt.xlabel('this is x')
plt.ylabel('this is y')
```

```python
plt.plot(t, s, 'r--', label='aaaa')
plt.plot(t*2, s, 'b--', label='bbbb')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is a demox')
plt.legend()
```
