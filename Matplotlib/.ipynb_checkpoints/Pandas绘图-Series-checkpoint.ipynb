```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s = Series(np.arange(1,6))
s.cumsum()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

kind : str
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    
s1.plot(kind='line', ax=None, figsize=None, use_index=True, title=None, grid=None, legend=False, style=None, logx=False, logy=False, loglog=False, xticks=None, yticks=None, xlim=None, ylim=None, rot=None, fontsize=None, colormap=None, table=False, yerr=None, xerr=None, label=None, secondary_y=False, **kwds)

```python
s1.plot(kind='line', grid=True, label='S1', legend=True, title='This is Series', \
        style='--')
s2.plot(label='S2')
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
```

```python
fig, ax = plt.subplots(2,1)
s1[0:10].plot(kind='bar', ax=ax[0], label='S1', legend=True, style='b')
s2.plot(ax=ax[1], label='S2', legend=True, style='r-')
```
