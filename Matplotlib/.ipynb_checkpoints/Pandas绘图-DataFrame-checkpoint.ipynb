```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
df = DataFrame(
    np.random.randint(1,10,40).reshape(10,4),
    columns=['A','B','C','D']
)
df
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df.iloc[5].plot(kind='pie', legend=True)
```

```python
for i in df.index:
    df.iloc[i].plot(label=str(i),legend=True)
```

```python
df.T.plot()
```
