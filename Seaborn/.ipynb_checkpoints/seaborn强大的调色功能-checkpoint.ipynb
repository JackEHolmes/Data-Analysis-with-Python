```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def sinplot():
    x = np.linspace(0,14,100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+%s)*(%s+0.75)'%(i,i))
    plt.legend()
```

```python
sinplot()
```

```python
pal_style = ['deep','muted','pastel','bright','dark','colorblind']
```

```python
for i in pal_style:
    sns.palplot(sns.color_palette(i))
```

```python
sns.set()
with sns.color_palette(pal_style[5]):
    sinplot()
```
