```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series, DataFrame
%matplotlib inline
```

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

# 1.axes_style and set_style

```python
style = ['darkgrid', 'dark', 'white', 'whitegrid', 'ticks']
```

```python
sns.set_style(style[4], {'axes.grid': True,'grid.color': 'b','grid.linestyle': '--'})
sinplot()
```

```python
sns.axes_style()
```

```python
sns.set()
sinplot()
```
